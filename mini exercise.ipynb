```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
response = requests.get('https://web-scraping-demo.zgulde.net/people')
html = response.text
html
```

```python
soup = BeautifulSoup(html)
soup
```

```python
articles = soup.select('.grid.grid-cols-2.gap-x-12.gap-y-16')
articles
```

```python
articles = articles[0].select('.person.border.rounded.px-3.py-5.grid.grid-cols-2.gap-x-3.bg-purple-50')
```

```python
def process_article(article):
    name  = article.select('h2')[0]
    tag = article.select('p')[0]
    email = article.select('.grid.grid-cols-9')[0].select('p')[0]
    phone = article.select('.grid.grid-cols-9')[0].select('p')[1]
    address = article.select('.address.grid.grid-cols-9')[0].select('p')[0]
    return {
        'name': name.text,
        'tag': tag.text,
        'email': email.text,
        'phone': phone.text,
        'address': address.text
    }

df = pd.DataFrame([process_article(article) for article in articles])
```
